```python
import pandas as pd
```

```python
df1 = pd.read_excel('./2020.xlsx', sheet_name = 1)
```

```python
df1.head()
```

```python
df_long = pd.melt(df1, id_vars = ['时间','month','合计'], var_name = '价格区间', value_name = '价格值')
```

```python
df_long.head()
```

```python
df_sort = df_long.loc[:, ['时间','month','价格区间','价格值','合计']]
```

```python
df_sort.head()
```

```python
df_sort.loc[:, '时间'] = pd.to_datetime(df_sort.loc[:,'时间'], format = 'Y%-%m-%d').dt.date
```

```python
df_sort.head()
```

```python
df_sort.to_excel('有品配送.xlsx')
```
